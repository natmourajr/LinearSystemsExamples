```python
# Adicionando as bibliotecas
import numpy as np
from scipy import signal
import control
```

```python
# Definindo um sistema

# control.tf([numerador,denominador])
# Ex: H(s) = (s-1) /(s^2 + 5s + 6) 
sys1 = control.tf([1,1], [1, 5, 6])
print sys1
```

```python
# Definindo uma entrada para simular a resposta do sistema
t = np.linspace(0,10,100)
x = np.cos(2*t)

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))

l1 = plt.plot(t,x,color=[0,0,1],
              linewidth=2.5,
              linestyle='solid',
              label='Entrada')

plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot de Entrada",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando a resposta ao impulso unitário
t = np.linspace(0,10,100)
T, y = control.impulse_response(sys1, t)

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))


l2 = plt.plot(t,y,color=[1,0,0],
              linewidth=2.5,
              linestyle='solid',
              label='Saida')


plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot da Resposta ao Impulso Unitario",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando a resposta ao degrau unitário
t = np.linspace(0,10,100)
T, y = control.step_response(sys1, t)

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))


l2 = plt.plot(t,y,color=[1,0,0],
              linewidth=2.5,
              linestyle='solid',
              label='Saida')


plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot da Resposta ao Degrau Unitario",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando a resposta a entrada zero
t = np.linspace(0,10,100)
T, y = control.initial_response(sys1, t,[1,1])

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))


l2 = plt.plot(t,y,color=[1,0,0],
              linewidth=2.5,
              linestyle='solid',
              label='Saida')


plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot da Resposta a Entrada Zero",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando a resposta ao estado zero
t = np.linspace(0,10,100)
x = np.cos(2*t)
T, y, x0 = control.forced_response(sys1, t, x,[0,0])

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))

l1 = plt.plot(t,x,color=[0,0,1],
              linewidth=2.5,
              linestyle='solid',
              label='Entrada')

l2 = plt.plot(t,y,color=[1,0,0],
              linewidth=2.5,
              linestyle='solid',
              label='Saida')


plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot da Resposta ao Estado Zero",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando a resposta completa
t = np.linspace(0,10,100)
x = np.cos(2*t)
T, y, x0 = control.forced_response(sys1, t, x,[1,1])

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig1 = plt.figure(figsize=(10,6))

l1 = plt.plot(t,x,color=[0,0,1],
              linewidth=2.5,
              linestyle='solid',
              label='Entrada')

l2 = plt.plot(t,y,color=[1,0,0],
              linewidth=2.5,
              linestyle='solid',
              label='Saida')


plt.grid(True)
plt.xlabel("Tempo (segundos)",fontsize=18,fontweight='bold')
plt.ylabel("Amplitude",fontsize=18,fontweight='bold')
plt.title("Plot da Resposta Completa",fontsize=18,fontweight='bold')
plt.legend()
```

```python
# Simulando o diagrama de bode

mag, phase, omega = control.bode(sys1,dB=True)

import matplotlib.pyplot as plt
%matplotlib inline  

plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15

plt.rc('legend',**{'fontsize':15})
plt.rc('font', weight='bold')

fig, ax = plt.subplots(figsize=(20,20),nrows=2, ncols=1)

ax[0].semilogx(omega, mag,
               color=[0,0,1], 
               linewidth=2.5,
               linestyle='solid',
               label='Sistema1'
              )
ax[0].grid(True)
ax[0].set_xlabel("Frequencia (Hz)",fontsize=18,fontweight='bold')
ax[0].set_ylabel("Amplitude (dB)",fontsize=18,fontweight='bold')
ax[0].set_title("Diagrama de Bode da Magnitude",fontsize=18,fontweight='bold')
ax[0].legend()


ax[1].semilogx(omega, phase,
               color=[0,0,1], 
               linewidth=2.5,
               linestyle='solid',
               label='Sistema1'
              )
ax[1].grid(True)
ax[1].set_xlabel("Frequencia (Hz)",fontsize=18,fontweight='bold')
ax[1].set_ylabel('Graus (o)',fontsize=18,fontweight='bold')
ax[1].set_title("Diagrama de Bode da Fase",fontsize=18,fontweight='bold')
ax[1].legend()
```
